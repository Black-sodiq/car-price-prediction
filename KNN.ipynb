```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('camry.csv',names=['Price','Condition','Mileage','Age','Year'])
```

```python
data.head()
```

```python
X = data.iloc[:, 1:5].values;
y = data.iloc[:, 0].values;
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
reg = KNeighborsRegressor(n_neighbors=5)
```

```python
reg.fit(X_train, y_train)
```

```python
print("test prediction:\n{}".format(reg.predict(X_test)))
```

```python
print("test score:\n{:.2f}".format(reg.score(X_test, y_test)))
```

```python
print("training set score:{:.2f}".format(reg.score(X_train, y_train)))
```
