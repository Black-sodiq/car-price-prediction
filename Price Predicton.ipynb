```python
datatest['seller_type'] = le.fit_transform(datatest['seller_type'])
datatest['country_of_use'] = le.fit_transform(datatest['country_of_use'])
datatest['weblog_date'] = le.fit_transform(datatest['weblog_date'])
datatest['car_name'] = le.fit_transform(datatest['car_name'])
datatest['car_brand'] = le.fit_transform(datatest['car_brand'])
datatest['date_of_advert '] = le.fit_transform(datatest['date_of_advert'])
datatest['last_advert_online '] = le.fit_transform(datatest['last_advert_online'])
datatest['car_type'] = le.fit_transform(datatest['car_type'])
datatest['gas_type'] = le.fit_transform(datatest['gas_type'])
datatest['gear_type'] = le.fit_transform(datatest['gear_type'])
datatest['Status_of_car_repair'] = le.fit_transform(datatest['Status_of_car_repair'])
datatest['model'] = le.fit_transform(datatest['model'])


datatest.head()
```

```python
datatest = datatest.drop(columns="date_of_advert")
datatest.head(n=2)
```

```python
datatest = datatest.drop(columns="last_advert_online")
datatest.head(n=2)
```

```python
X_train = datatrain.loc[:, datatrain.columns != 'naira_price']
y_train = datatrain.loc[:, 'naira_price']
X_test = datatest

print(X_train.shape, y_train.shape, X_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=20, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
print("training set score:{:.2f}".format(regressor.score(X_train, y_train)))
print("test set score:{:.2f}".format(regressor.score(X_test, y_pred)))
```
